```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importing datasets
movie_data_path = 'https://github.com/keithvtls/AIDA2-CpE---AIML/blob/main/Correlation/movie_metadata_cleaned.csv?raw=true'
```

```python
# reading csv file including pandas
df_movie = pd.read_csv(movie_data_path)
```

```python
df_movie
```

```python
## Number 1

X = df_movie.director_facebook_likes.values
y = df_movie.gross.values

def find_cor(X,y):
    ## Finding the mean of X and y
    x_bar = X.mean()
    y_bar = y.mean()

    ## get the difference for each data point from their means
    sst_x = X-x_bar
    sst_y = y-y_bar

    ## Finding the covariance of X and y
    cov_xy = np.cov(X,y, bias=True)[0,1]

    ## Finding the standard deviation of X and y
    std_X = np.std(X)
    std_y = np.std(y)

    ## Finding the correlation
    R = cov_xy /(std_X*std_y)
    return R

## Plotting the figure
plt.figure(figsize=(10,10))
plt.grid()
plt.scatter(X,y)
plt.xlabel('GROSS') 
plt.ylabel('LIKES') 
plt.show()


find_cor(X,y)
```

```python
i = df_movie.plot_keywords.str.count('\|') +1 
i = i.replace(1,0).values
j = df_movie.budget



## Plotting the figure
plt.figure(figsize=(10,10))
plt.grid()
plt.scatter(i,j)
plt.ylabel('BUDGET') 
plt.xlabel('PLOT COUNT') 
plt.show()


find_cor(i,j)
```
